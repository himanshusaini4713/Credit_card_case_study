The user should be able to specify the product (Gold/Silver/Platinum) and time period (yearly or monthly) and the function should automatically take these inputs while identifying the top 10 customers.

```python
cust_spend_data
```

```python
cust_spend_data['Year']=cust_spend_data.Month.apply(lambda x:x.year)
cust_spend_data['Monthly']=cust_spend_data.Month.apply(lambda x:x.month)
```

```python
def top10Customers(product_category,time_period):
    return cust_spend_data.loc[(cust_spend_data.Product == product_category)].groupby(['Customer','City','Product',time_period]).Amount.sum().reset_index().sort_values('Amount',ascending=False).head(10)
```
